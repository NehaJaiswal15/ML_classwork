```python
import numpy as np
import random
```

```python
names = ["Anuj", "Aman", "Abhay", "Anjali", "Akshat", "Amaya", "Ayush", "Amit", "Anunay", "Ajay", "Avishka", "Anushka", "Ankur", "Amrit", "Ankush", "Ashish", "Abhinav", "Alia", "Ankita", "Aishi" ]
```

```python
students = []
```

```python
for i in range(20):
    roll = i+1
    marks1 = random.randint(40, 100)
    marks2 = random.randint(40, 100)
    total = marks1 + marks2

    students.append([roll, names[i], marks1, marks2, total])
    
```

```python
np.savetxt("student.csv", students, fmt="%s", delimiter=",", header = "Roll,Name,Marks1,Marks2,Total", comments="")
```

```python
data = np.genfromtxt("student.csv", delimiter=",", dtype=str)
```
